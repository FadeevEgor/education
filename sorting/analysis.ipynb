```python
import numpy as np
import pandas as pd

import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib

font = {'size'   : 18}


matplotlib.rc('font', **font)
```

```python
N = 5000
titles = ["comparisons", "swaps"]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
for i in range(2):
    df = pd.read_csv(f"./data/{N}_{titles[i]}.csv", index_col=0)
    sns.lineplot(data=df, ax=axs[i])
    axs[i].set_xlabel("reversed <-  random  -> sorted")
    axs[i].set_title(titles[i])

plt.tight_layout()
```
