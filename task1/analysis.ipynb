```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import seaborn as sns
import matplotlib
matplotlib.rc("font", **{"size":18})
```

```python
langs = ["ru", "zh"]
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

for i, l in enumerate(langs):
    df = pd.read_csv(f"data/{l}/substr_8_measure.csv", index_col=0)
    df.index = np.log10(df.index)
    # print(df)
    sns.lineplot(data=df, ax=axs[i])
    axs[i].set_title(l)
    
    
plt.tight_layout()
```

```python
text_size = [(i, 10**i) for i in range(3, 9)]


langs = ["ru", "zh"]
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))
for j, l in enumerate(langs):
    naive = np.zeros(len(text_size))
    rk = np.zeros(len(text_size))
    for i, (e, N) in enumerate(text_size):
        df = pd.read_csv(f"data/{l}/substr_{e}_measure.csv", index_col=0)
        naive[i], rk[i] = df.loc[100]
        
    df = pd.DataFrame({"Naive Search": naive, "Rabin Karp": rk}, index=[x[1] for x in text_size] )
    sns.lineplot(data = df, ax=axs[j])
    
    
```
