```python
import os

import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
folder = "./measurements/"
tables = os.listdir(folder)
tables
```

```python
dfs = []
for t in tables:
    path = os.path.join(folder, t)
    dfs.append(pd.read_csv(path, index_col=["nodes", "edges"]))
    
df = pd.concat(dfs, axis="index")
df
```

```python
sns.set(font_scale=2)
def plot(df, nodes):
    fig, axs = plt.subplots(ncols=2, figsize=(10, 8))
    fig.suptitle(f"{nodes} nodes")
    sns.lineplot(data=df[["calls", "calls_total"]], ax=axs[0])
    sns.lineplot(data=df["routes"], ax=axs[1])
    for ax in axs:
        ax.xaxis.set_ticks(df.index)
    fig.tight_layout()
    plt.show()
    
```

```python
plot(df.loc[3], 3)
```

```python
plot(df.loc[4], 4)
```
